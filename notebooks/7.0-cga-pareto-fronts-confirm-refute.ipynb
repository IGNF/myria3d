```python
confirm_study_path = "/var/data/cgaydon/data/202110_building_val/logs/multiruns/2021-11-19_15-34-01/optimization_results.yaml" #M5V1.6
refute_study_path = "/var/data/cgaydon/data/202110_building_val/logs/multiruns/2021-11-19_15-34-31/optimization_results.yaml"
paths = [confirm_study_path, refute_study_path]
```

```python
import yaml
import seaborn as sns; sns.set();
import numpy as np
import matplotlib.pyplot as plt

for path in paths:
    stream = open(path, "r")
    docs = yaml.safe_load(stream)
    metrics = np.array([s["values"] for s in docs["solutions"]])
    sns.scatterplot(x=metrics[:,0], y=metrics[:,1])
    plt.title("Fronts de Pareto des solutions")
    plt.xlabel("Taux de d√©cision")
    plt.ylabel("Justesse")

plt.legend(["Confirmation", "Refutation"])
```
